```python
import pandas as pd 
   
# Dataframe generation 
string = 'PandasisFun'
llist = 3 * [pd.Series(list(string))] 
df = pd.DataFrame(data = llist)  #converting list of series into a dataframe
df.head()
```

```python
print("Original dataframe:\n" +       df.to_string(index = False, 
      header = False), end = '\n\n') #viewing orginal Data
```

```python
#using apply function we are sorting values between columns for each row
df.apply(lambda x:x.sort_values(ascending=False),axis=1) 
```

```python
#using applymap function we can access each element of each cell in dataframe and apply some operation
df.applymap(lambda x:x.upper()+'%') 
```

```python
#using map function we can achieve some functionality on any column of datframe which is a series
df[0]=df.iloc[:,0].map(lambda x: x.replace('P','Z'))
df.head()
```
